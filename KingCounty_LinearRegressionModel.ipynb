### Adding Crime Data

```python
kingcounty_crime = pd.read_csv('/users/christianmoya/Documents/Flatiron/Phase_2/dsc-phase-2-project/kingcounty_crime.csv')
kingcounty_crime.set_index('zip', inplace=True)
kingcounty_crime.head()
```

```python
def propcrime(zipcode):
    for key in range(len(kingcounty_crime)):
            propertycrime = kingcounty_crime['propertycrime_percapita'][zipcode]
    return propertycrime

propcrime_percapita = []
for z in df['zip']:
    try:
        crime = propcrime(z)
    except:
        crime = 0 
    propcrime_percapita.append(crime)
    
propcrime_percapita
```

```python
def violcrime(zipcode):
    for key in range(len(kingcounty_crime)):
            violentcrime = kingcounty_crime['violentcrime_percapita'][zipcode]
    return violentcrime

violcrime_percapita = []
for z in df['zip']:
    try:
        crime = violcrime(z)
    except:
        crime = 0 
    violcrime_percapita.append(crime)
    
violcrime_percapita
```

```python
df['prop_crime'] = propcrime_percapita
df['viol_crime'] = violcrime_percapita
```

```python
df['totalcrime_percapita'] = df['prop_crime'] + df['viol_crime']
df.head()
```

```python
# Remove comma from data and convert into int
def convert_to_int(value):
    x = value.replace(',','')
    return int(x)

total_students = []
for val in processed_data2['total_student']:
    tot_stud = convert_to_int(val)
    total_students.append(tot_stud)
```
