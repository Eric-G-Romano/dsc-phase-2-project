```python
df= df.drop(columns=['yr_renovated'])
print(df.isnull().sum())
```

```python
df['waterfront'].fillna(value= 0.0, inplace=True)
print(df.isnull().sum())
```

```python
df['bathrooms'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title('Number of bathrooms')
plt.ylabel('Number of houses')
plt.xlabel('bathrooms')
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(df['price'], df['bathrooms'])
plt.title('Scatter plot between price and bathrooms')
ax.set_xlabel('price')
ax.set_ylabel('bathrooms')
plt.show()
```

```python
con_feat = ['price', 'sqft_lot', 'sqft_above', 'sqft_basement', 'bedrooms', 'bathrooms', 'floors']
cat_feat = [ 'waterfront', 'condition', 'grade']
```

```python
dummy = pd.get_dummies(df[cat_feat], drop_first=True)
```

```python
df = pd.concat([df, dummy], axis = 1)
df = df.drop([ 'waterfront', 'condition', 'grade'])
```

```python
plt.figure(figsize=[15,15])
sns.heatmap(df.corr(),annot=True);
```
